```python
from sklearn.linear_model import LinearRegression #불러오기
import statsmodels.api as sm

#데이터 세팅
y_data = df_insure['charges']
x_data = df_insure.drop(['charges', 'child_yn'], axis = 1)
print(x_data.columns)

#dummy variable 만들기
X = df_insure[['sex', 'smoker' , 'region']]
X = pd.get_dummies(data = X, drop_first = True)

# 만들어진 dummy variable 합치기
x_data2 = x_data[['age', 'bmi','children']].join(X, how = 'left')
x_data2

#상수항 결합
x_data3 = sm.add_constant(x_data2)
ls = sm.OLS(y_data, x_data3).fit()
print(ls.summary())
```

> Adj. R-squared : 0.749

### 2.Simple Data Exploration(data = scores.csv, question.csv)

#### 2.1 변수탐색

##### 1. 학생별로 4번의 시험점수를 평균을 내어 이를 scoreboard로 저장하라.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#데이터 저장
score = pd.read_csv('c:/python/package/2018_2/week1/scores.csv', index_col=0)
score

#평균내어 저장하자
scoreboard = pd.concat([score,(score.drop(['student'], axis = 1).mean(axis = 1))], axis = 1)
scoreboard.columns = ['student', 'first', 'second', 'third', 'fourth', 'avg']
scoreboard
```

##### 2. question.csv를 이용하여 scoreboard에 학생 별 질문 횟수가 기록된 n_question이라는 새로운 열을 생성하라. 이후, 학생 별로 계산한 평균점수에 학생 별 질문횟수를 더해 최종 점수를 계산한 totalscore이라는 새로운 열을 생성하라

```python
question = pd.read_csv('c:/python/package/2018_2/week1/question.csv', index_col=0)
question

#데이터를 조인하자
scoreboard_2 = scoreboard.join(question, on = 'student', how = 'left')
scoreboard_2['totalscore'] = scoreboard_2['avg'] + scoreboard_2['n_question']
scoreboard_2
```

##### 3. totalscore를 기준으로 상위 30%까지 A, 상위 60%까지 B, 상위 90%까지 C, 나머지는 F. grade라는 열에 이 기준에 의거하여 학점을 부여하라

```python
quan = pd.Series(scoreboard_2['totalscore'].quantile([0.7, 0.4, 0.1]))
quan.index = [0,1,2]
quan
```

```python
def grade_maker(tt) :
    if tt > quan[0] : 
        return 'A'
    elif tt > quan[1] : 
        return 'B'
    elif tt > quan[2] : 
        return'C'
    else : 
        return 'F'
    
scoreboard_2['grade'] = scoreboard_2.apply(lambda x: grade_maker(x['totalscore'])
                                           , axis = 1)
scoreboard_2
```

##### 4. A,B,C,F 학점별로 시험 별 평균점수의 차이가 있었는데 시각화해보기 위해 다음을 수행하라

- 학생 별 학점을 grades라는 이름으로 scores에 포함시키고
- 다음에 해당하는 데이터 프레임인 scores_gg를 생성하라. 학점 별로 1~4차 mean_first부터 mean_fourth를 낸 다음 key를 test, value를 score로 지정하여 하나의 열로 정리한 데이터 프레임
- scores_gg안의 test열에 있는 mean_first를 1,,, mean_fourth를 4로 바꾼 다음 이 열을 factor처리. 그리고 다음 시각화를 수행

```python
scores_gg = scoreboard_2.groupby('grade').mean()[['first', 'second' ,'third', 'fourth']]
scores_gg_t = scores_gg.T
scores_gg_t
```

```python
x = np.arange(1,5)
aa = scores_gg_t['A']
bb = scores_gg_t['B']
cc = scores_gg_t['C']
ff = scores_gg_t['F']


line1, = plt.plot(x, aa, '-o', label = 'A')
line2, = plt.plot(x, bb, '-o', label = 'B')
line3, = plt.plot(x, cc, '-o', label = 'C')
line4, = plt.plot(x, ff, '-o', label = 'F')

plt.grid(True)

plt.xticks([1,2,3,4]) #x축의 값들을 원하는 값으로 바꿀 수 있다.

plt.show()
```
